### Remove Empty Rate Value

```python
def getTypes(rate):
    types = type(rate)
    return types

df['rateType'] = df['rate'].apply(getTypes)

df.head()
```

```python
df['rate'].unique()
```

```python
df.groupby('rate')['rate'].count()
```

```python
df[df['rate'].isnull()].head()
```

```python
def replaceEmptyValue(rate):
    if isinstance(rate, str):
        rate = rate.strip()
        if rate != 'NEW' and rate != '-':
            actualRate = rate.split('/')[0].strip()
            return float(actualRate)
        else:
            return 0.0
    elif isinstance(rate, float):
        return 0.0
    
df['rate'] = df['rate'].apply(replaceEmptyValue)
```
