### Remove Empty Rate Value

```python
def getTypes(rate):
    types = type(rate)
    return types

df['rateType'] = df['rate'].apply(getTypes)

df.head(1)
```

```python
df['rate'].unique()
```

```python
df.groupby('rate')['rate'].count()
```

```python
df[df['rate'].isnull()].head(1)
```

```python
def replaceEmptyValue(rate):
    if isinstance(rate, str):
        rate = rate.strip()
        if rate != 'NEW' and rate != '-':
            actualRate = rate.split('/')[0].strip()
            return float(actualRate)
        else:
            return 0.0
    elif isinstance(rate, float):
        return 0.0
    
df['rate'] = df['rate'].apply(replaceEmptyValue)
```

```python
df = df.drop(['rateType'], axis=1)
```

```python
df['rate'].unique()
```

```python
approx_cost = df[df['approx_cost(for two people)'].isnull()]
approx_cost.head(1)
```

```python
df.dropna(subset=['rest_type', 'approx_cost(for two people)', 'cuisines'], inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df['cuisines_count'] = [len(cui) for cui in df['cuisines'].str.split(',')]
df['dish_liked_count'] = [len(dish) if dish is not np.NAN else 0 for dish in df['dish_liked'].str.split(',')]
df.head(1)
```
